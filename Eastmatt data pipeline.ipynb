```python
import pandas as pd
import os

# -----------------------------
# 1. FILES AND MONTH MAPPING
# -----------------------------
files = {
    "SEPERSLEEK  DATA JULY.xlsx": "July",
    "SUPERSLEEK DATA AUG.xlsx": "August",
    "SUPERSLEEL DATA SEP.xlsx": "September",
    "SUPERSLEEK DATA OCT.xlsx": "October",
    "SUPERSLEEK DATA NOV.xlsx": "November",
    "SUPERSLEEK  DATA DEC.xlsx": "December",
    "SUPERSLEEK DATA JAN.xlsx": "January"
}

dataframes = []

# -----------------------------
# 2. READ, ADD MONTH, COMBINE
# -----------------------------
for file, month in files.items():
    df = pd.read_excel(file, sheet_name= "Table1")
    df["Month"] = month
    dataframes.append(df)

df = pd.concat(dataframes, ignore_index=True)

# Ensure text columns are strings and lowercase for matching
df["Description"] = df["Description"].astype(str).str.lower()
df["Sub category"] = df["Sub category"].astype(str).str.lower()

# -----------------------------
# 3. CLEANING LOGIC
# -----------------------------

# --- BLEACH LOGIC ---
# If Sub category is bleach AND description contains color keywords → Bleach Colours
bleach_colour_keywords = ["col", "colour", "colours", "colors"]

mask_bleach_colour = (
    (df["Sub category"] == "bleach") &
    (df["Description"].str.contains("|".join(bleach_colour_keywords), regex=True))
)

df.loc[mask_bleach_colour, "Sub category"] = "bleach colours"

# Bleach under fabric softener → regular bleach
mask_bleach_fabric = (
    (df["Sub category"] == "fabric softener") &
    (df["Description"].str.contains("bleach"))
)

df.loc[mask_bleach_fabric, "Sub category"] = "bleach"

# --- HAND WASH STANDARDIZATION ---
df.loc[
    df["Sub category"].isin(["hand wash", "handwash"]),
    "Sub category"
] = "hand wash"

# --- SHOWER GEL STANDARDIZATION ---
df.loc[
    df["Sub category"].isin(["showel gel", "shower gel", "body wash"]),
    "Sub category"
] = "shower gel"

# --- TOILET BLOCK STANDARDIZATION ---
df.loc[
    df["Sub category"].isin(["toilet block", "toilet blocks"]),
    "Sub category"
] = "toilet block"

# --- WATER PURIFIERS ---
df.loc[
    df["Description"].str.contains("aquaguard"),
    "Sub category"
] = "water purifiers"

# -----------------------------
# 4. OPTIONAL: TITLE CASE FOR OUTPUT
# -----------------------------
df["Sub category"] = df["Sub category"].str.title()
df["Month"] = df["Month"].str.title()

# -----------------------------
# 5. SAVE CLEANED DATA
# -----------------------------
df.to_excel("SUPERSLEEK BRANCH FINAL.xlsx", index=False)

print("Data successfully combined and cleaned!")
```
